
# Control Flow


## Conditionals — Basic `if`

```python

age = 16
if age < 18:
    print("Minor")
```

## Conditionals — `if / else`

```python

age = 20
if age < 18:
    print("Minor")
else:
    print("Adult")
```

## Conditionals — `if / elif / else`

```python

age = 70
if age < 18:
    print("Minor")
elif age < 65:
    print("Adult")
else:
    print("Senior")
```

## Conditionals — Nested `if`

```python

age = 25
is_student = True

if age < 30:
    if is_student:
        print("Young student")
    else:
        print("Young professional")
else:
    print("Adult")
```

## Conditionals — Logical operators

```python

score = 85
attendance = 90

if score > 80 and attendance > 75:
    print("Eligible for certificate")
else:
    print("Not eligible")
```

## Loops — `for` over list

```python

fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

## Loops — `for` with range

```python

for i in range(5):
    print(i)
```

## Loops — `while` loop

```python

count = 0
while count < 5:
    print(count)
    count += 1
```

## Loops — `break`

```python

for i in range(10):
    if i == 5:
        break
    print(i)
```

## Loops — `continue`

```python

for i in range(5):
    if i == 3:
        continue
    print(i)
```

## Pattern — Right-angled triangle

```python

rows = 5
for i in range(1, rows + 1):
    for j in range(i):
        print("*", end=" ")  # stay on the same line
    print()
```

## Pattern — Inverted triangle

```python

rows = 5
for i in range(rows, 0, -1):
    for j in range(i):
        print("*", end=" ")
    print()
```

## Pattern — Pyramid

```python

rows = 5
for i in range(1, rows + 1):
    print(" " * (rows - i), end="")
    for j in range(i):
        print("* ", end="")
    print()
```

## Pattern — Diamond

```python

rows = 5

for i in range(1, rows + 1):
    print(" " * (rows - i), end="")
    for j in range(i):
        print("* ", end="")
    print()

for i in range(rows - 1, 0, -1):
    print(" " * (rows - i), end="")
    for j in range(i):
        print("* ", end="")
    print()
```

## Pattern — Hollow square (uses if/else)

```python

rows = 5
for i in range(rows):
    for j in range(rows):
        if i == 0 or i == rows - 1 or j == 0 or j == rows - 1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

## Pattern — Hollow pyramid (complex conditionals)

```python

rows = 5
for i in range(1, rows + 1):
    print(" " * (rows - i), end="")
    for j in range(1, 2 * i):
        if j == 1 or j == 2 * i - 1 or i == rows:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```
